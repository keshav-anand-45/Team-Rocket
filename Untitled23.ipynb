```python
import numpy as np
import glob
import struct
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, welch
```

```python
FS = 2_000_000
DATA_PATH = r"C:\Users\HP\Downloads\data\data\*.txt"
MAX_SAMPLES = 300000
```

```python
# SAFE IQ READER
def read_iq_stream(fname, max_samples):

    import numpy as np
    import struct
    import re

    I = []
    Q = []

    with open(fname) as f:

        for line in f:

            if len(I) >= max_samples:
                break

            # extract hex numbers from line
            vals = re.findall(r'[0-9A-Fa-f]{8}', line)

            if len(vals) < 2:
                continue

            try:
                I.append(struct.unpack(">f", bytes.fromhex(vals[0]))[0])
                Q.append(struct.unpack(">f", bytes.fromhex(vals[1]))[0])
            except:
                continue

    return np.array(I) + 1j*np.array(Q)

```

```python
# LOWPASS
def lowpass(sig,cutoff):

    b,a = butter(5,cutoff/(FS/2))
    return lfilter(b,a,sig)

```

```python
# LOAD FILE
files = sorted(glob.glob(DATA_PATH))

print("Total files:",len(files))

f = files[20]   # file inside signal region
print("Testing file:",f)

iq = read_iq_stream(f,MAX_SAMPLES)

print("Samples loaded:",len(iq))
```

```python
# SPECTROGRAM
plt.figure(figsize=(8,4))
plt.specgram(iq, NFFT=1024, Fs=FS)
plt.title("Spectrogram")
plt.show()
```

```python
# CARRIER DETECTION
spectrum = np.fft.fftshift(np.fft.fft(iq))
freqs = np.fft.fftshift(np.fft.fftfreq(len(iq),1/FS))
power = np.abs(spectrum)**2
carrier = freqs[np.argmax(np.abs(spectrum))]
print("Detected Carrier (Hz):",carrier)
```

```python
power_db = 10*np.log10(power)

window = 200
power_smooth = np.convolve(power_db, np.ones(window)/window, mode='same')

plt.figure(figsize=(10,4))
plt.plot(freqs, power_smooth)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power (dB)")
plt.title("Smoothed Power Spectrum")
plt.grid()
plt.show()
```

```python
# --- SHARP + ZOOMED SPECTRUM ---

# convert to dB safely
power_db = 10*np.log10(power + 1e-12)

# smooth a little
window = 200
power_smooth = np.convolve(power_db, np.ones(window)/window, mode='same')

# zoom window around carrier
zoom = 100000   # 100 kHz around carrier

mask = (freqs > carrier-zoom) & (freqs < carrier+zoom)

plt.figure(figsize=(10,4))

plt.plot(freqs[mask]/1e3, power_smooth[mask], linewidth=2)

plt.title("Zoomed Carrier Spectrum")
plt.xlabel("Frequency (kHz)")
plt.ylabel("Power (dB)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(freqs, power)
plt.title("Frequency vs Power Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.grid()
plt.show()
```

```python
# BASEBAND
t = np.arange(len(iq))/FS
baseband = iq*np.exp(-1j*2*np.pi*carrier*t)

filtered = lowpass(baseband,50000)
```

```python
# SPECTRUM
f_spec,P = welch(filtered,FS,nperseg=4096)

plt.figure(figsize=(8,4))
plt.plot(f_spec,10*np.log10(P))
plt.title("Signal Spectrum (Bandwidth)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power (dB)")
plt.show()
```

```python
# CONSTELLATION
plt.figure(figsize=(6,6))
plt.scatter(np.real(filtered[:20000]), np.imag(filtered[:20000]), s=1)
plt.xlabel("I")
plt.ylabel("Q")
plt.title("Constellation")
plt.show()
```

```python
def costas_loop(signal):

    phase = 0
    freq = 0

    alpha = 0.01
    beta = 0.0001

    out = []

    for s in signal:

        corrected = s*np.exp(-1j*phase)

        error = np.sign(corrected.real)*corrected.imag

        freq += beta*error
        phase += freq + alpha*error

        out.append(corrected)

    return np.array(out)
```

```python
tracked = costas_loop(filtered)

plt.figure(figsize=(6,6))
plt.scatter(np.real(tracked[:20000]), np.imag(tracked[:20000]), s=1)
plt.title("Constellation After Carrier Recovery")
plt.xlabel("I")
plt.ylabel("Q")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

env = np.abs(tracked)

# remove DC
env = env - np.mean(env)

# autocorrelation
corr = np.correlate(env, env, mode='full')
corr = corr[len(corr)//2:]

# normalize
corr = corr / np.max(corr)

# convert lag to time
lags = np.arange(len(corr))
times = lags / FS

# ignore first peak
search = corr[int(FS/50000):int(FS/2000)]

peak = np.argmax(search)

symbol_period = (peak + int(FS/50000)) / FS

symbol_rate = 1 / symbol_period

print("Estimated Symbol Rate:", symbol_rate)
```

```python
plt.figure(figsize=(8,4))
plt.plot(times[:2000], corr[:2000])
plt.title("Autocorrelation for Symbol Timing")
plt.xlabel("Time lag (seconds)")
plt.ylabel("Correlation")
plt.show()
```

```python
sr = int(symbol_rate)
sps = int(FS / sr)

print("Samples per symbol:", sps)
```

```python
import numpy as np
import glob
import struct
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

FS = 2_000_000
DATA_PATH = r"C:\Users\HP\Downloads\data\data\*.txt"
MAX_SAMPLES = 200000
```

```python
# SAFE IQ READER
def read_iq_stream(fname, max_samples):

    import re

    I = []
    Q = []

    with open(fname) as f:

        for line in f:

            if len(I) >= max_samples:
                break

            vals = re.findall(r'[0-9A-Fa-f]{8}', line)

            if len(vals) < 2:
                continue

            try:
                I.append(struct.unpack(">f", bytes.fromhex(vals[0]))[0])
                Q.append(struct.unpack(">f", bytes.fromhex(vals[1]))[0])
            except:
                continue

    return np.array(I) + 1j*np.array(Q)
```

```python
# LOWPASS FILTER
def lowpass(sig,cutoff):

    b,a = butter(5,cutoff/(FS/2))
    return lfilter(b,a,sig)

```

```python
# COSTAS LOOP
def costas_loop(signal):

    phase = 0
    freq = 0

    alpha = 0.01
    beta = 0.0001

    out = []

    for s in signal:

        corrected = s*np.exp(-1j*phase)

        error = np.sign(corrected.real)*corrected.imag

        freq += beta*error
        phase += freq + alpha*error

        out.append(corrected)

    return np.array(out)
```

```python
# LOAD FILE LIST
files = sorted(glob.glob(DATA_PATH))
print("Total files:",len(files))


# -------------------------
# USE SIGNAL REGION (5-70)
# -------------------------
iq_all = []

for f in files[5:70]:

    print("Loading:",f)

    iq = read_iq_stream(f,MAX_SAMPLES)

    iq_all.append(iq)

iq_all = np.concatenate(iq_all)

print("Total samples:",len(iq_all))

```

```python
# SPECTROGRAM (opt)
plt.figure(figsize=(8,4))
plt.specgram(iq_all,NFFT=1024,Fs=FS)
plt.title("Spectrogram")
plt.show()
```

```python
# CARRIER DETECTION
spectrum = np.fft.fftshift(np.fft.fft(iq_all))
freqs = np.fft.fftshift(np.fft.fftfreq(len(iq_all),1/FS))

carrier = freqs[np.argmax(np.abs(spectrum))]

print("Detected Carrier (Hz):",carrier)
```

```python
# BASEBAND CONVERSION
t = np.arange(len(iq_all))/FS

baseband = iq_all*np.exp(-1j*2*np.pi*carrier*t)
```

```python
# LOWPASS
filtered = lowpass(baseband,50000)
```

```python
# COSTAS LOOP
tracked = costas_loop(filtered)
```

```python
# CONSTELLATION
plt.figure(figsize=(6,6))
plt.scatter(np.real(tracked[:20000]),np.imag(tracked[:20000]),s=1)
plt.title("Constellation After Carrier Recovery")
plt.show()
```

```python
# SYMBOL RATE
sr = 41667
sps = int(FS/sr)

print("Symbol Rate:",sr)
print("Samples per symbol:",sps)
```

```python
# MATCHED FILTER
h = np.ones(sps)/sps
mf = lfilter(h,1,tracked)
```

```python
# SYMBOL EXTRACTION
symbols = mf[sps//2::sps]
```

```python
# SYMBOL CONSTELLATION
plt.figure(figsize=(6,6))
plt.scatter(np.real(symbols[:5000]),np.imag(symbols[:5000]),s=2)
plt.title("Symbol Constellation")
plt.show()
```

```python
# BPSK DEMODULATION
bits = (np.real(symbols) > 0).astype(int)
```

```python
# CCSDS SYNC WORD
sync_hex = 0x1ACFFC1D
sync_bits = np.array([int(b) for b in format(sync_hex,"032b")])

diff_bits = np.mod(np.diff(bits),2)

streams = [bits,1-bits,diff_bits,1-diff_bits]



for data in streams:

    corr = np.correlate(data*2-1,sync_bits*2-1,mode="valid")

    print("Max correlation:",np.max(corr))

    idx = np.argmax(corr)

    if corr[idx] > 20:

        print("\nSYNC FOUND")
        print("Bit index:",idx)
        break
```

```python
corr = np.correlate(bits*2-1, sync_bits*2-1, mode="valid")

sync_positions = np.where(corr > 18)[0]

print("Total packets detected:", len(sync_positions))

packet_number = 1

for start in sync_positions:

    packet = bits[start:start+1024]

    byte_vals = []

    for i in range(0, len(packet), 8):

        byte = packet[i:i+8]

        if len(byte) < 8:
            break

        val = int("".join(map(str, byte)), 2)

        byte_vals.append(val)

    header = byte_vals[:6]
    payload = byte_vals[6:]

    print("\n----------------------------------")
    print("Packet", packet_number)
    print("Bit position:", start)
    print("Header:", header)
    print("Payload start:", payload[:20])

    packet_number += 1
```

```python
env = np.abs(tracked)

env = env - np.mean(env)

spectrum = np.abs(np.fft.fft(env))
freqs = np.fft.fftfreq(len(env), 1/FS)

mask = freqs > 0
freqs = freqs[mask]
spectrum = spectrum[mask]

# ignore very low frequencies
mask = freqs > 1000
freqs = freqs[mask]
spectrum = spectrum[mask]

symbol_rate = freqs[np.argmax(spectrum)]

print("Estimated Symbol Rate:", symbol_rate)
plt.figure(figsize=(8,4))
plt.plot(freqs, spectrum)
plt.title("Envelope Spectrum (Symbol Rate Peak)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.show()
```
